## Step 2: Data Cleaning Process

- Removing Irrelevant columns.

```python
df.drop(columns =['Unnamed: 0'], inplace= True)
```

- Missing Data Detection

```python
df.isnull().sum()
```

Observed Missing Columns:-
- Bedrooms
- Bathrooms

## Missing Data Treatment

```python
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].mean())
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].mean())
```

```python
df.isna().sum()
```

## Date Formatting.

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date'].dtype
```

## Step 3: Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Q1.What is the distrubution of house prices?

```python
plt.figure()
plt.hist(df['price'], bins=30)
plt.title("Distrubution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

### Analysis :- 
- Distrubution is Right Skewed.
- Majority houses are in tha mid price range

### Conclusion.
- Luxury houses are fewer, market is domainated by affordable properties.

### Q2. How does living area affect price?

```python
plt.scatter(df['sqft_living'], df['price'])
plt.xlabel("Sqft of Living")
plt.ylabel("Price")
plt.title("Price vc Living Area")
plt.show()
```

## Analysis :- 
- Strong positive correlation.
## Conclusion.
- Bigger homes generally cost more.

## Q3. What is the relationship between bedrooms and Price.

```python
sns.boxplot(x = 'bedrooms', y='price', data=df)
plt.title("Bedrooms vc Price")
plt.show()
```

## Analysis :- 
- Price increases till certain bedroom count.

## Conclusion:-
- After 6 bedrooms, price groth slows.

## Q4. Do bathrooms impack house price?

```python
sns.scatterplot(x ='bathrooms', y ='price', data = df)
plt.title("Price vc Bathrooms")
plt.show()
```

## Analysis :- 
- Strong impact than bathrooms.
## Conclusion:-
- Bathrooms are more important than bedrooms for pricing.

## Q5.Does waterfront view increase price?

```python
sns.boxplot(x='waterfront', y='price', data=df)
plt.title("Waterfront vs Price")
plt.show()
```

## Analysis
- Waterfront houses are much costlier
## Conclusion
- Waterfront location is a premium feature.

## Q6.What is the correlation among key variables?

```python
corr = df[['price','sqft_living','bathrooms','bedrooms','grade']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## Analysis
- sqft_living has strongest correlation
## Conclusion
- Living space is the best price predictor.
